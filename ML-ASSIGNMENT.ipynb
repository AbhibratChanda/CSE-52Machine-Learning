```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 7, 9, 11])

X = X.reshape(-1, 1)
X_b = np.c_[np.ones((X.shape[0], 1)), X]
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(theta[0])
print(theta[1])
y_pred = X_b.dot(theta)
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 6, 8, 10])
y = np.array([5, 9, 12, 14, 19])

n = len(x)
mean_x = np.mean(x)
mean_y = np.mean(y)

numerator = np.sum((x - mean_x) * (y - mean_y))
denominator = np.sum((x - mean_x)**2)
slope = numerator / denominator
intercept = mean_y - slope * mean_x
print(f"Intercept: {intercept}")
print(f"Slope: {slope}")
y_pred = intercept + slope * x
plt.scatter(x, y, color='green', label='Observed data')
plt.plot(x, y_pred, color='orange', label=f'Line: y = {intercept:.2f} + {slope:.2f}x')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
x_value = 15
predicted_y = intercept + slope * x_value
print(f"Prediction for x = {x_value}: y = {predicted_y}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([1, 3, 5])
y = np.array([4, 8, 14])

X = np.column_stack((np.ones(len(x1)), x1))
beta = np.linalg.inv(X.T @ X) @ (X.T @ y)
beta_0, beta_1 = beta
print(f"Intercept (beta_0): {beta_0}")
print(f"Coefficient for x1 (beta_1): {beta_1}")
y_pred = X @ beta
plt.scatter(range(len(y)), y, color='green', label='Actual values (y)')
plt.plot(range(len(y)), y_pred, color='orange', label='Predicted values (y)')
plt.xlabel('Index')
plt.ylabel('y')
plt.title('Linear Regression: Actual vs Predicted')
plt.legend()
plt.grid(True)
plt.show()
x1_new = 8
y_new = beta_0 + beta_1 * x1_new
print(f"Prediction for x1 = {x1_new}: y = {y_new}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6])
x2 = np.array([3, 6, 9])
y = np.array([5, 10, 15])

X = np.column_stack((np.ones(len(x1)), x1, x2))

beta = np.linalg.inv(X.T @ X) @ (X.T @ y)

beta_0, beta_1, beta_2 = beta

print(f"Intercept (beta_0): {beta_0}")
print(f"Coefficient for x1 (beta_1): {beta_1}")
print(f"Coefficient for x2 (beta_2): {beta_2}")

y_pred = X @ beta

plt.scatter(range(len(y)), y, color='blue', label='Actual data (y)')
plt.plot(range(len(y)), y_pred, color='red', label='Predicted values (y)')
plt.xlabel('Index')
plt.ylabel('y')
plt.title('Multiple Linear Regression: Actual vs Predicted')
plt.legend()
plt.grid(True)
plt.show()

x1_new = 7
x2_new = 10
y_new = beta_0 + beta_1 * x1_new + beta_2 * x2_new
print(f"Prediction for x1 = {x1_new}, x2 = {x2_new}: y = {y_new}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6])
x2 = np.array([3, 6, 9])
y = np.array([5, 10, 15])

x1_bar = np.mean(x1)
x2_bar = np.mean(x2)
y_bar = np.mean(y)

numerator_a1 = np.sum((x1 - x1_bar) * (y - y_bar))
denominator_a1 = np.sum((x1 - x1_bar) ** 2)
a1 = numerator_a1 / denominator_a1

numerator_a2 = np.sum((x2 - x2_bar) * (y - y_bar))
denominator_a2 = np.sum((x2 - x2_bar) ** 2)
a2 = numerator_a2 / denominator_a2
a0 = y_bar - a1 * x1_bar - a2 * x2_bar
print(f"Intercept (a0): {a0}")
print(f"Slope for x1 (a1): {a1}")
print(f"Slope for x2 (a2): {a2}")
y_pred = a0 + a1 * x1 + a2 * x2
plt.scatter(range(len(y)), y, color='blue', label='Actual data (y)')
plt.plot(range(len(y)), y_pred, color='red', label=f'Predicted values (y)')
plt.xlabel('Index')
plt.ylabel('y')
plt.title('Multiple Linear Regression: Actual vs Predicted')
plt.legend()
plt.grid(True)
plt.show()

x1_new = 7
x2_new = 10
y_new = a0 + a1 * x1_new + a2 * x2_new
print(f"Prediction for x1 = {x1_new}, x2 = {x2_new}: y = {y_new}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.array([1, 2, 3, 4, 5])
y_vals = np.array([2, 4, 5, 4, 5])

m = 0  
b = 0  

alpha = 0.1

iterations = 5
history_m = [m]
history_b = [b]

for i in range(iterations):
    y_pred = m * x_vals + b
    
    dm = (-2 / len(x_vals)) * np.sum(x_vals * (y_vals - y_pred))  
    db = (-2 / len(x_vals)) * np.sum(y_vals - y_pred)  
    
    m -= alpha * dm
    b -= alpha * db
    
    history_m.append(m)
    history_b.append(b)

print(f"Final slope (m): {m}")
print(f"Final intercept (b): {b}")

plt.scatter(x_vals, y_vals, color='blue', label='Data points')

x_line = np.linspace(0, 6, 100)
for i in range(len(history_m)):
    plt.plot(x_line, history_m[i] * x_line + history_b[i], label=f"Iteration {i}")

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Gradient Descent for Linear Regression (5 iterations)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.rand(100) * 10
y = 2 * X + 1 + np.random.randn(100)

def cost_function_linear(X, y, m, b):
    n = len(X)
    y_pred = m * X + b
    cost = (1 / (2 * n)) * np.sum((y_pred - y) ** 2)
    return cost

m_values = np.linspace(-5, 5, 100)
b = 1
cost_values = [cost_function_linear(X, y, m, b) for m in m_values]
plt.figure(figsize=(8, 6))
plt.plot(m_values, cost_values, label="Cost Function (MSE)", color='b')
plt.title("Cost Function for Linear Regression")
plt.xlabel("Slope (m)")
plt.ylabel("Cost (MSE)")
plt.grid(True)
plt.legend()
plt.show()
```
